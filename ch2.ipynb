# ii. Produce a scatterplot matrix of the first ten columns or variables of the data.

A scatterplot matrix is just a set of scatterplots for each pair in the cartesian product of features.

Looks like you can't actually do a scatterplot matrix very easily with matplotlib(?), but I guess pandas has a function.

That function is documented here: https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.plotting.scatter_matrix.html#pandas.plotting.scatter_matrix

```python
from pandas.plotting import scatter_matrix
scatter_matrix(college.iloc[:,1:11])
```

Wow that looks bad. Let's try again.

```python
scatter_matrix(college.iloc[:,1:11], figsize=(12, 12))
```

### iii. Boxplots of outstate vs private.

### iv. New category *elite*

```python
college['Elite'] = 'no'
```

```python
college.loc[college['Top10perc']>50, 'Elite'] = 'yes'
```

```python
college.head()
```

```python
college.loc[college['Elite']=='yes'].describe()
```

```python
college.boxplot(column='Outstate', by='Elite')
```
