```python
import numpy as np
import pandas as pd
```

```python
smkt = pd.read_csv('data/Smarket.csv', usecols=range(1,10))
smkt.head()
```

```python
smkt.corr()
```

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
formula = 'Direction ~ Lag1 + Lag2 + Lag3 + Lag4 + Lag5 + Volume'
model = smf.glm(formula=formula, data=smkt, family=sm.families.Binomial()).fit()
model.summary()
```

```python
modpred = model.predict()
```

```python
smkt['Prediction'] = 'Up'
```

```python
smkt.loc[modpred > 0.5, 'Prediction'] = 'Down'
```

```python
smkt.head(20)
```

```python
pd.crosstab(smkt.Direction, smkt.Prediction)
```

```python
model2 = smf.glm(formula=formula, data=smkt.loc[smkt.Year<2005], family=sm.families.Binomial()).fit()
model2.summary()
```

```python
mod2pred = model2.predict(smkt.loc[smkt.Year==2005])
mod2pred.tail()
```

```python
smkt['Prediction2'] = 'Up'
smkt.loc[(smkt.Year == 2005) & (mod2pred > 0.5), 'Prediction2'] = 'Down'
smkt.tail()
```

```python
pd.crosstab(smkt.loc[smkt.Year==2005].Direction, smkt.loc[smkt.Year==2005].Prediction2)
```
